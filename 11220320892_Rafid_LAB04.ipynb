```python
import matplotlib.pyplot as plt
import numpy as np

A = (0, 6)
B = (5, 0)
C = (0, 20)
D = (20, 0)

plt.plot(*A, "yo", label="A(0, 6)")
plt.plot(*B, "bo", label="B(5, 0)")
plt.plot(*C, "yo", label="C(0, 20)")
plt.plot(*D, "bo", label="D(20, 0)")

plt.plot([A[0], B[0]], [A[1], B[1]], 'brown', label="12x + 10y = 60")
plt.plot([C[0], D[0]], [C[1], D[1]], 'gray', label="4x + 4y = 80")

x = np.linspace(0, 20, 100)
y_line1 = (60 - 12 * x) / 10
y_line2 = (80 - 4 * x) / 4

plt.fill_between(x, np.minimum(y_line1, y_line2), 0, where=(y_line1 >= 0) & (y_line2 >= 0), color="blue", alpha=0.5)

plt.xticks(np.arange(0, 30, step=5))
plt.yticks(np.arange(0, 30, step=5))
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Feasible region of the Linear Programming Problems")
plt.grid(True)
plt.axis()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

A = (0, 8)
B = (2, 0)
C = (0, 15)
D = (9, 0)
E = (0, 9)
F = (10, 0)

plt.plot(*A, "purple", label="A(0, 8)")
plt.plot(*B, "green", label="B(2, 0)")
plt.plot(*C, "blue", label="C(0, 15)")
plt.plot(*D, "cyan", label="D(9, 0)")
plt.plot(*E, "yellow", label="E(0, 9)")
plt.plot(*F, "red", label="F(10, 0)")

plt.plot([A[0], B[0]], [A[1], B[1]], "brown", label="8x + 2y = 16")
plt.plot([C[0], D[0]], [C[1], D[1]], "green", label="5x + 3y = 45")
plt.plot([E[0], F[0]], [E[1], F[1]], "gray", label="9x + 10y = 90")

x = np.linspace(0, 10, 400)
y1 = (16 - 8 * x) / 2
y2 = (45 - 5 * x) / 3
y3 = (90 - 9 * x) / 10

plt.fill_between(x, np.maximum(y1, 0), np.minimum(np.minimum(y2, y3), 10), where=(y1 <= y2) & (y1 <= y3) & (y2 >= 0) & (y3 >= 0), color="lightblue", alpha=0.5)

plt.xticks(np.arange(0, 15, step=1))
plt.yticks(np.arange(0, 20, step=1))
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.title("Feasible region for Linear Programming problems")
plt.grid(True)
plt.show()
```
